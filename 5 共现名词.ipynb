# 合并版

```python
import pickle,json,time
with open('output/chongyanlist.pkl', 'rb') as file:
    cy_set = pickle.load(file)
print("已经导入重言词列表,长度：", len(cy_set))
with open('output/set_qts.pkl', 'rb') as file:
    loaded_set = pickle.load(file)
print("已经导入全唐诗,长度：", len(loaded_set))

dictoken = {}
dicpos = {}
for word in cy_set:
    # 寻找待分词的句子
    toposli = []
    for i in loaded_set:
        if word in i:
            toposli.append(i)
    print(f'待匹配词语{word}',len(toposli),toposli)

    #分词和词性标注
    result = HanLP.parse("。".join(toposli), tasks='pos/pku')
    # print(result)

    # 分别存储分词结果和词性标注结果
    tokenli = []
    posli = []
    for line in result["tok/fine"]:
        tokenli += line

    for line in result["pos/pku"]:
        posli += line

    print("分词列表",tokenli)
    print("词性列表",posli)

    # 存储到字典中
    dictoken[word] = tokenli
    dicpos[word] = posli

    # print(dicpos[word])
    time.sleep(2)

file_path = 'output/cytokenjson.json'
with open(file_path,'w',encoding='utf-8') as f:
    json.dump(dictoken,f)
file_path = 'output/cyposjson.json'
with open(file_path,'w',encoding='utf-8') as f:
    json.dump(dicpos,f)
```

```python
npos = ['n','nr','ns','nx','nz','Ng']
nworddic = {}
from collections import Counter
for x in cy_set:
    nli = []
    tokenli = dictoken[x]
    posli = dicpos[x]
    for ind,pos in enumerate(posli):
        if pos in npos:
            word = tokenli[ind]
            nli.append(word)
    # 使用 Counter 统计每个元素的个数
    element_count = Counter(nli)
    element_count_sorted = element_count.most_common()
    # print(element_count_sorted)
    nworddic[x] = element_count_sorted
    print(x,element_count_sorted)
# print(nworddic)
```

```python
cy_posjson = []
for i in cy_set:
    adic = {}
    adic['word'] = i
    adic['token'] = dictoken[i]
    adic['pos'] = dicpos[i]
    adic['n'] = nworddic[i]
    cy_posjson.append(adic)
    print(adic)
file_path = 'output/cyposjson_process.json'
with open(file_path,'w',encoding='utf-8') as f:
    json.dump(cy_posjson,f)
```
