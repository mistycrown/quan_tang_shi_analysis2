```python
# 被查老師坑了一波。之前查老師用的re.finditer,不知道哪裏錯了，反正數據就是不對20230707
import re

def extract_context(text, word):
    # 使用正则表达式找到所有匹配的上下文
    pattern = r'(?:.{0,20})' + word + r'(?:.{0,20})'
    matches = re.findall(pattern, text)

    return matches

# 测试文本
text = "这是一段示例文本，用于演示如何提取包含指定字的上下文。这段文本是中文的示例文本。"

# 匹配包含字的上下文
word = "示例"
contexts = extract_context(text, word)

# 打印结果
for context in contexts:
    print(context)
```

```python
# 20230708 改进版，可输入多个词语
import os
import re
import pandas as pd


bookdic = {'mao_shi':'毛诗','chu_ci':'楚辭','wen_xuan':'文選',"yu_tai_xin_yong":'玉台新詠','le_fu_shi_ji':"樂府詩集",'quan_tang_shi':'全唐詩'}
def bookfind(book,wordli):
    numli = []
    path = os.path.join(r"static\corpus",book)
    #print(path)
    filelist = os.listdir(path)
    #print(filelist)
    for x in wordli:
        num = 0
        for i in filelist:
            with open(os.path.join(path,i),"r",encoding="utf-8") as f:
                txt = f.read()
                contexts = extract_context(txt,x)
                for j,context in enumerate(contexts):
                    context = context.replace(x,f'[{x}]')
                    print(f'{i}-{j}-{x}',context)
                num += len(contexts)
        numli.append(num)
    return numli
```

```python
wordli =["黯惨",'黯黪']#,'瀚瀚',"瀚漫","寂寥"
data = []
for k in bookdic:
    print("========={}==========".format(bookdic[k]))
    numli = bookfind(k,wordli)
    #print(k,numli)
    data.append(numli)

df = pd.DataFrame(data,columns=wordli,index=['诗经','楚辭','文選','玉台新詠','樂府詩集','全唐詩'])
print(df)
```

# 查找二元組

```python
# 尋找一個單字經常搭配的二元組
import re
from collections import Counter

def find_word_bigrams(corpus, word):
    pattern = r'[\u4e00-\u9fa5]'
    words = re.findall(pattern, corpus)

    # 构建二元组列表
    bigrams = [(words[i], words[i+1]) for i in range(len(words)-1) if word in (words[i], words[i+1])]

    # 统计二元组出现次数
    counter = Counter(bigrams)

    # 按出现次数排序并返回结果列表
    sorted_bigrams = sorted(counter.items(), key=lambda x: x[1], reverse=True)
    return sorted_bigrams

# 测试语料库
# corpus = "这是一个没有分词的中文文本。我们需要对它进行分词处理。"

with open(r'output\total_corpus.txt','r',encoding='utf-8') as f:
    corpus = f.read()
```

```python
# 查找包含字的二元组并按出现次数排序
word = '森'
result = find_word_bigrams(corpus, word)

# 打印结果列表
num = 0
for bigram, count in result:
    print(f"{bigram}: {count}")
    num +=1
    if num >10 :
        break
```
